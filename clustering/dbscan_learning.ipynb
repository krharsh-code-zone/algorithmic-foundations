calculating Eps

```python
# finding Eps
Eps = 0
for i in range(100):
  dist = np.sqrt((x[i] - x)**2)
  dist = list(dist.sum(axis=1))
  eps = 0
  for _ in range(6):
    eps += min(dist)
    dist.remove(min(dist))
  eps = eps/6
  # print(eps)
  Eps += eps/100
Eps = Eps - (Eps/3)
Eps = round(Eps, 2)
print(Eps)
```

```python
# 0: not visited
# 1: visited
visit = [0]*100
status = [0]*100
MinPts = 6
```

```python
def within(data,pi):
  '''
  returns a array of indices which lies within the distancce of Eps to the given p: point
  requires  data: whole 2D data
            point: index of point
  '''
  dist = np.sqrt(((x[pi] - x)**2).sum(axis=1))
  temp = []
  for i in range(len(dist)):
    if dist[i]<= Eps:   #removed visited list updating block from the if block
      temp.append(i)
  temp = np.array(temp)
  return temp

print(within(x,9))
```

```python
temp = set()
cluster_no = 1
def visited(x,pi,temp,visit,status,cluster_no,MinPts):
  visit[pi] = 1
  status[pi] = cluster_no
  if len(within(x,pi)) >= MinPts:  #core
    for i in within(x,pi):
      if visit[i] == 0 or visit[i] == -1: # non-visited
        status[i] = cluster_no
        temp.add(i)
        visited(x,i,temp,visit,status,cluster_no,MinPts)
  return temp
print(visited(x,9,temp,visit,status,cluster_no,MinPts))
```

```python
cluster_no = 0
for i in range(len(x)):
  if visit[i] == 0:
    if len(within(x,i)) >= MinPts: #core
      cluster_no += 1
      print(visited(x,i,temp,visit,status,cluster_no,MinPts))
    else: #noise
      visit[i] = -1
      status[i] = -1
print(len(visit), visit)
print(len(status), status)
```
